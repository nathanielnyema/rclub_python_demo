## Some Examples with the Iris Dataset
### first we load in the data
**Note**: Here we are using sklearn's interface for loading an existing dataset. The object this returns is not quite like any typical data structure you would usually work with in Python outside of machine learning  contexts. As a result, we will try to get the data into a pandas dataframe which many visualization libraries are built to use as inputs. to do this. Lets first get a feel for what's in the dataset. These sklearn dataset objects have a method called keys() which allows us to see what attributes we can access are.

```python
iris = datasets.load_iris()
iris.keys()
```

### create a pandas dataframe

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head(10)
```

When accessing these dataframes we can use the column name as an index as follows


```python
iris_df['sepal length (cm)']
```

The result is an object called a pandas Series. We can use this same syntax to create new columns. Lets add the targets to the dataframe

```python
iris_df['target'] = iris.target
iris_df.head(10)
```

If we don't have spaces in the column name we can alternatively access the column as if we were accessing the attribute of an object

```python
iris_df.target
```

When we start plotting it will be useful for the target column entries to actually be the names of the targets instead of the number labels. To do this we can use a method of the pandas Series called `replace()`. To use this method we need to create a dictionary, the built-in datatype in Python for enumerated collections. This just means we can explicitly define keys for the multiple values we store in this variable. We will use this to map each label $0,1$ or $2$ to the corresponding name setosa, versicolor, or virginica.

```python
# when we start plotting it will be use fu
target_labels = {
    0: iris.target_names[0],
    1: iris.target_names[1],
    2: iris.target_names[2]
}
#a more pythonic way
target_labels = dict(enumerate(iris.target_names))
iris_df['target'] = iris_df.target.replace(target_labels)
iris_df.head(10)
```

### now let's plot!

```python
# I first want to get a list of all possible combinations of features then we will try plotting all pairs
# in scatter plots 
from itertools import combinations
combos = list(combinations(iris.feature_names,2))
print(combos)
```

```python
%matplotlib inline
# we first need to create the axes to make our plots on
# the following code creates an array of axes with 2 rows
# and (# of combos)/2 columns. The // operator forces the division
# to return an integer

fig, ax = plt.subplots(2, len(combos)//2, figsize = (12,6))
for i, combo in enumerate(combos):
    x, y = combo # unpacking the tuple
    if i == ax.shape[-1]-1:
        # if we're on the edge of the array
        # i want to show a legend
        # we need to specify how we want to
        # show the legend if we want to show
        # one. in this case we want to show the full legend
        show_legend = 'full'
    else:
        # otherwise don't bother. set show_legend to a False boolean
        show_legend = False
        
    #use seaborn's built in scatter plot function
    sns.scatterplot(data=iris_df, # a dataframe in longform
                    x = x, # the column with the x data
                    y = y, # the column with the y data
                    legend = show_legend, # whether or not to show the legend
                    palette = "Set2", # a color map to draw the colors from (see: https://matplotlib.org/3.5.0/tutorials/colors/colormaps.html)
                    hue = 'target', # a column corresponding to a grouping variable indicating what datapoints should be the same color
                    ax = ax.flatten()[i] # the axis to plot this on)
                   ) 

sns.despine() # remove the top and right spines of the plot
fig.tight_layout(pad = 1.)
```

### what about plotting in 3D ??

```python
%matplotlib qt
ax = plt.axes(projection='3d')
ax.set_prop_cycle(color = sns.color_palette('Set2', 3))
features = [ 'sepal width (cm)', 'petal width (cm)', 'petal length (cm)']

for i in iris_df.target.unique():
    x,y,z = iris_df.loc[iris_df.target == i, features].values.T
    ax.scatter3D(x,y,z)
ax.set_xlabel(features[0]);
ax.set_ylabel(features[1]);
ax.set_zlabel(features[2]);
```

## Time Series Data: FMRI

```python
%matplotlib inline
#this time we are using seaborn to load the data in so it's already a dataframe!
fmri = sns.load_dataset("fmri")
fmri.head(10)
```

```python
# the line plot function is mostly the same as scatterplot!
# we don't necessarily need to specify the axis if we are just
# making one plot

sns.lineplot(data = fmri, x = 'timepoint', y='signal', 
             hue = 'event', palette='mako', style = 'region')
sns.despine()
```

### just for fun let's try using PCA to view this data in a different way
PCA requires the data to be an array of n_samples x n_variables. To get this we'll do a bit of reshaping of the dataframe. 

```python
fmri_reshaped = fmri.set_index(['subject', 'region', 'event', 'timepoint'])
fmri_reshaped.head(10)
```

```python
fmri_reshaped = fmri_reshaped.unstack('timepoint').sort_index()
cue_idx = fmri_reshaped.index.get_level_values("event") == 'cue'
fmri_reshaped.head(10)
```

```python
fmri_arr = fmri_reshaped.values
fmri_arr
```

```python
%matplotlib qt
from sklearn.decomposition import PCA
# run PCA and get the first
pcs = PCA(n_components = 3).fit_transform(fmri_reshaped.values)

# make a 3D PCA plot
ax = plt.axes(projection='3d')
ax.scatter3D( pcs[cue_idx, 0],  pcs[cue_idx,1],  pcs[cue_idx,2])
ax.scatter3D( pcs[~cue_idx,0], pcs[~cue_idx,1],  pcs[~cue_idx,2])
ax.set_xlabel('PC1');
ax.set_ylabel('PC2');
ax.set_zlabel('PC3');
```
